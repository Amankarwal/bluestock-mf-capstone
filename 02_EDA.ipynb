```python
import matplotlib.pyplot as plt

sector_weight = holdings.groupby(
    'sector'
)['weight_pct'].sum().sort_values(
    ascending=False
)

sector_weight.head(10).plot(
    kind='bar',
    figsize=(10,5)
)

plt.title("Top 10 Sectors by Portfolio Weight")
plt.ylabel("Weight %")
plt.show()
```

```python
top_holdings = holdings.groupby(
    'stock_name'
)['weight_pct'].mean().sort_values(
    ascending=False
).head(10)

top_holdings.plot(
    kind='bar',
    figsize=(10,5)
)

plt.title("Top 10 Holdings")
plt.ylabel("Weight %")
plt.show()
```
